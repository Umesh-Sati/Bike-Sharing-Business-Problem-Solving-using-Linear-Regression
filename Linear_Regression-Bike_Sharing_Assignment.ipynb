## let's start the Data Preprocessing


```python
#creating the dummy variables for mnth, weekday, weathersit and season variables:

months     = pd.get_dummies(df.mnth,drop_first=True)
weekdays   = pd.get_dummies(df.weekday,drop_first=True)
weathersit = pd.get_dummies(df.weathersit,drop_first=True)
seasons    = pd.get_dummies(df.season,drop_first=True)

#concating the original dataframe with the dummy variable dataframe:

df_new = pd.concat([df, months, weekdays, weathersit, seasons], axis=1)

#dropping original columns as it is not needed in model building process:

df_new.drop(['season','mnth','weekday','weathersit'], axis = 1, inplace = True)

#Checking head of the new dataframe:

df_new.head()
```

# Now split the dataset into Test & Train Dataset 

```python
# splitting the dataset:

np.random.seed(0)
df_train, df_test = train_test_split(df_new, train_size = 0.7, test_size = 0.3, random_state = 1000)
```

```python
# Let's check the head and shape of train dataset:

print(df_train.shape)
df_train.head()
```

```python
#rescaling the features using MinMaxScaler to normalize the data:

scaler = MinMaxScaler()

#normalizing all the countinuous variables:

df_train[num_cols] = scaler.fit_transform(df_train[num_cols])

#checking head of the data:

df_train.head()
```

```python
# Now describing the data:

df_train.describe()
```

```python
# Checking the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (27, 27))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.title("Correlation coefficient",fontsize=27)
plt.xticks(rotation = 60, fontsize = 14)
plt.yticks(fontsize = 14)
plt.show()
```

Target variable cnt follows:

- There is a positive correlation with variables yr & temp.
- Negative correlation with month jan & feb and season spring.


# Now let's build a linear regression model 

```python
#splitting the df_train in X,y sets:

y_train = df_train.pop('cnt')
X_train = df_train
```

```python
#using recursive feature elimination (RFE) for initial feature selection:

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, n_features_to_select= 15)
rfe = rfe.fit(X_train, y_train)

#Checking the list of all 15 features:

list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# Checking the selected features names:

features = X_train.columns[rfe.support_]
print(features)
```

```python
# Calculating VIF(Variance Inflation Factor) by creating a function:

def VIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif
```

```python
# Updating the X_train with RFE selected columns:

X_train_rfe = X_train[features]
```

# Let's build the first model:

```python
#Adding a constant as it is neccesary in case of statsmodel :

X_train_lm = sm.add_constant(X_train_rfe)

#Fitting the model:

lr_1 = sm.OLS(y_train, X_train_lm).fit()

#Checking the summary of the models:

print(lr_1.summary())
```

```python
#Checking VIF:
VIF(X_train_rfe)
```

The atemp colum needs to be dropped as its insignificant(high p-value) and highly correlated(high VIF value).

```python
# Let's drop the atemp column:

X_train_new = X_train_rfe.drop("atemp", axis = 1)

#checking head of the data:

X_train_new.head()
```

# Let's build the second model:

```python
#Adding a constant:

X_train_lm = sm.add_constant(X_train_new)

#Fitting the model:

lr_2 = sm.OLS(y_train, X_train_lm).fit()

#Checking the summary:

print(lr_2.summary())
```

```python
#Checking the VIF:

VIF(X_train_new)
```

The `hum` column needs to be dropped as it is highly correlated(high VIF value).

```python
#Dropping the column hum:

X_train_new = X_train_new.drop("hum", axis = 1)

#Checking the head of the data:

X_train_new.head()
```

# Let's build the Third model:

```python
#add the constant:

X_train_lm = sm.add_constant(X_train_new)

#fitting the model:

lr_3 = sm.OLS(y_train, X_train_lm).fit()

#Checking the summary of the model:

print(lr_3.summary())
```

```python
#Checking the VIF values:

VIF(X_train_new)
```

We are droping the workingday column as it is highly correlated(high VIF value).

```python
#dropping the workingday column:

X_train_new = X_train_new.drop("workingday", axis = 1)

#Checking the head of the data:

X_train_new.head()
```

# Let's build the Fourth model:

```python
#add constant:

X_train_lm = sm.add_constant(X_train_new)

#fitting the model:

lr_4 = sm.OLS(y_train, X_train_lm).fit()

#checking the summary of the model:

print(lr_4.summary())
```

```python
#Checking the VIF:

VIF(X_train_new)
```

We are removing the `sat` feature as insignificantit is (high p-value).

```python
#dropping the sat column:

X_train_new = X_train_new.drop("sat", axis = 1)

#checking the head of the data:

X_train_new.head()
```

# Let's build the fifth model: 

```python
#add constant:

X_train_lm = sm.add_constant(X_train_new)

#fitting the model:

lr_5 = sm.OLS(y_train, X_train_lm).fit()

#Checking the summary of the model:

print(lr_5.summary())
```

```python
#Checking the VIF:

VIF(X_train_new)
```

### 'Since the VIF values are less than 5 and p-value are less than 0.05. Thus, it looks like all the variables are significant and have low multicollinearity. Now we move to the next step to make predictions using this model'.

# Let's do the Residual analysis on the train data:

### First assumption 

```python
#predicting the target variable on train data using the final model:

y_train_pred = lr_5.predict(X_train_lm)

#Visualizing the residauls:

plt.figure(figsize = (10,5))
sns.distplot((y_train - y_train_pred,), bins = 20)
plt.title("Error Terms", fontsize = 15)
plt.xlabel("Error", fontsize = 10)
plt.show()
```

- From the above histogram we can see that errors are normally distributed.
- Mean of the Error terms is 0.

### Second assumption

```python
# Using the visualizing the error terms using scatter plot to check if there is any visible pattern:

residuals = (y_train - y_train_pred)

plt.figure(figsize =(10,5))
plt.scatter(y_train,residuals)
plt.plot(y_train,(y_train - y_train), '-r')
plt.ylabel("y_train")
plt.xlabel("Residuals")
plt.show()
```

- Since the visible pattern is not easily identifiable hence, we can say that error terms are independent of each other.
- It seems that error terms have constant variance(Homoscedasticity).

### Third assumption 

```python
#Let's check if there is any multicollinearity between the predictor variables:

VIF(X_train_new)
```

- Since all the VIF values are less than 5 so there is no multicollinearity b/w the predictor variables.

```python
#Checking how the model performs against the training set:

c = [i for i in range(0,len(X_train),1)]
plt.figure(figsize = (16,8))
plt.plot(c,y_train, color="blue")
plt.plot(c,y_train_pred, color="pink")
plt.title('y_Actual vs y_Predicted', fontsize = 20)
plt.xlabel('Index', fontsize = 18)
plt.ylabel('Demand', fontsize = 18)
plt.show()
```

- We can see that it looks good, it is predicting all the high and low almost accurately.

```python
#Visualizing the actual vs predicted values on train data:

fig = plt.figure(figsize = (7,4))
plt.scatter(y_train, y_train_pred)
fig.suptitle('y_train vs y_train_pred', fontsize = 15) 
plt.xlabel('y_train', fontsize = 15)
plt.ylabel('y_train_pred', fontsize = 15) 
plt.show()
```

```python
#Checking the R-Squared and adjusted R-Squared value on the actual vs predicted values of the train data set:

r_score_train = round(r2_score(y_train,y_train_pred),4)
print("R-Squared value:",r_score_train)

N = len(X_train_new) 
p = len(X_train_new.columns)
r2_train_adj = round((1-((1-r_score_train)*(N-1)/(N-p-1))),4)
print("adjusted R-Squared value:",r2_train_adj)
```

# Lets do the prediction using the final model:

```python
#Scaling the test data set:

df_test[num_cols] = scaler.transform(df_test[num_cols])

#Checking the head of the data:

df_test.head()
```

```python
#Describing the data:

df_test.describe()
```

```python
#dividing the test data set in X and y:

y_test = df_test.pop('cnt')
X_test = df_test
```

```python
X_test = X_test[X_train_new.columns]

#adding the constant in test data set:

X_test = sm.add_constant(X_test)

#Predicting using the model:

y_test_pred = lr_5.predict(X_test)
```

```python
#Checking the R-Squared and adjusted R-Squared value on the actual vs predicted values of the test data set:

r_score_test = round(r2_score(y_test,y_test_pred),4)
print("R-Squared value:",r_score_test)

N = len(X_test) 
p = len(X_train_new.columns)
r2_test_adj = round((1-((1-r_score_test)*(N-1)/(N-p-1))),4)
print("adjusted R-Squared value:",r2_test_adj)
```
